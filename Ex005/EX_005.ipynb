<p>Use the best edge detection methods, to find the watch numbers' edges, you can use denoising methods before detecting edges!. ( the output has to be so clear, without any noise!)</p>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_img_plt(img, c_map ='gray', fig_hight=8, fig_width=6, vmin=0, vmax=255):
    plt.figure(figsize=(fig_hight, fig_width), dpi=100)
    plt.axis('off')
    plt.imshow(img, cmap=c_map, vmin=vmin, vmax=vmax)
```

```python
def show_multi_img(images , titles ,  row , column , vmin=0 , vmax=255):
    fig = plt.figure(figsize=(15,17) , dpi=100)

    for i in range(len(images)):
        fig.add_subplot(row , column , i+1)
        plt.imshow(images[i] , cmap='gray' , vmin=vmin , vmax=vmax) 
        plt.axis('off')
        plt.title(titles[i])
```

```python
#read Image
img = cv2.imread(r'images/clock.png')
```

```python
# Color Conversion
img_gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
```

```python
# Denoising
img_blur = cv2.GaussianBlur(img_gray , (3,3) , 0)
```

```python
show_img_plt(img_blur)
```

## Sobel

```python
#Gradient
sobelx = cv2.Sobel(src = img_blur , ddepth=-1 , dx=1 , dy=0 , ksize=5)
sobely = cv2.Sobel(src = img_blur , ddepth=-1 , dx=0 , dy=1 , ksize=5)
```

```python
show_multi_img([sobelx , sobely],['sobelx' , 'sobely'] , 1 , 2)
```

```python
gradient = cv2.addWeighted(np.abs(sobelx) , 0.5 , np.abs(sobely) , 0.5 , 0)
```

```python
show_img_plt(gradient)
```

## laplacian

```python
laplacian = cv2.Laplacian(img_blur , -1)
```

```python
show_img_plt(laplacian)
```

## Canny

```python
canny = cv2.Canny(image = img_blur , threshold1=245 , threshold2=250)
```

```python
show_img_plt(canny)
```

```python
#read Image
output = cv2.imread(r'images/output.jpg')
```

```python
show_multi_img([gradient , laplacian , canny , output , img_blur],['gradient' , 'laplacian' , 'canny' , 'output' , 'img_blur'],1,5)
```
